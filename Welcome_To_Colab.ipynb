## Analyze Top 5 Customers

### Subtask:
Calculate the total transaction amount for each customer and display the top 5 customers with the highest transaction values, along with a bar chart visualization.


**Reasoning**:
The first step in analyzing the top 5 customers is to merge the transaction data (`sample_df`) with customer information (`customers_df`) to link transaction amounts with customer segments.



````
merged_df = pd.merge(sample_df, customers_df, on='CustomerID', how='left')

print("Merged DataFrame Head:")
print(merged_df.head())
````

**Reasoning**:
Now that the dataframes are merged, the next step is to group the merged DataFrame by 'CustomerID' and 'Segment', calculate the sum of 'Price' (total transaction amount), sort the results, and select the top 5 customers. Then, print the results as a table.



````
customer_total_transactions = merged_df.groupby(['CustomerID', 'Segment'])['Price'].sum().reset_index()
top_5_customers = customer_total_transactions.sort_values(by='Price', ascending=False).head(5)
top_5_customers = top_5_customers.rename(columns={'Price': 'Total Transaction Amount'})

print("Top 5 Customers by Total Transaction Amount:")
print(top_5_customers.to_string(index=False))
````

**Reasoning**:
To complete the subtask, the next step is to create a bar chart visualization of the top 5 customers by their total transaction amount. This involves using the `top_5_customers` DataFrame and `matplotlib` to plot the data with appropriate labels and title.



````
import matplotlib.pyplot as plt
import IPython.display as display
import io
import base64

# Create the bar chart
fig = plt.figure(figsize=(10, 6))
plt.bar(top_5_customers['CustomerID'].astype(str) + ' (' + top_5_customers['Segment'] + ')', top_5_customers['Total Transaction Amount'], color='skyblue')
plt.xlabel('Customer ID (Segment)')
plt.ylabel('Total Transaction Amount')
plt.title('Top 5 Customers by Total Transaction Amount')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Top 5 Customers by Total Transaction Amount"
display.display(display.Markdown(F"![{alt}]({image})"))
plt.close(fig)
````

## Best Transaction Time

### Subtask:
Determine the day of the week and the hour of the day when gas station users make the most transactions. Display the results in a clear format, possibly with a heatmap or bar chart visualization for clarity and include appropriate labels and legends.


**Reasoning**:
The first step is to extract the day of the week and the hour of the day from the existing 'Date' and 'Time' columns in the `sample_df`. This will involve converting the 'Date' column to datetime objects (if not already done in previous steps) and using datetime properties to get the day name and hour. A custom order for the days of the week will also be defined to ensure correct sorting in visualizations.



````
import pandas as pd

# Ensure 'Date' column is datetime type
sample_df['Date'] = pd.to_datetime(sample_df['Date'])

# Extract day of the week and hour of the day
sample_df['Day of Week'] = sample_df['Date'].dt.day_name()
sample_df['Hour'] = sample_df['Time'].apply(lambda x: int(x.split(':')[0]))

# Define a custom order for days of the week
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
sample_df['Day of Week'] = pd.Categorical(sample_df['Day of Week'], categories=day_order, ordered=True)

print("Sample DataFrame with new 'Day of Week' and 'Hour' columns:")
print(sample_df[['Date', 'Time', 'Day of Week', 'Hour']].head())
````

**Reasoning**:
The next step is to group the DataFrame by 'Day of Week' and 'Hour', then calculate the total 'Price' for each group to determine the transaction volume. This will provide the data needed for further analysis and visualization.



````
transaction_volume = sample_df.groupby(['Day of Week', 'Hour'])['Price'].sum().reset_index()

print("Transaction Volume by Day of Week and Hour:")
print(transaction_volume.head())
````

**Reasoning**:
Now that the transaction volume is grouped by day and hour, the next step is to reshape this data into a pivot table. This format is ideal for creating a heatmap, with days of the week as rows and hours as columns, showing the total transaction price for each combination.



````
transaction_volume_pivot = transaction_volume.pivot_table(index='Day of Week', columns='Hour', values='Price', fill_value=0)

print("Transaction Volume Pivot Table (for Heatmap):")
print(transaction_volume_pivot.head())
````

**Reasoning**:
The previous code block produced a `FutureWarning` due to changes in Pandas' `pivot_table` function when dealing with categorical data. I will explicitly set `observed=False` to maintain the current behavior and suppress the warning. After reshaping the data, the next step is to create a heatmap visualization of the transaction volume across different days and hours. This involves importing `seaborn` and `matplotlib.pyplot`, then using the `transaction_volume_pivot` to generate a heatmap with appropriate labels, title, and a color bar. Finally, the heatmap will be base64 encoded and displayed as a markdown image.



````
transaction_volume_pivot = transaction_volume.pivot_table(index='Day of Week', columns='Hour', values='Price', fill_value=0, observed=False)

print("Transaction Volume Pivot Table (for Heatmap):")
print(transaction_volume_pivot.head())

import matplotlib.pyplot as plt
import seaborn as sns
import IPython.display as display
import io
import base64

# Create the heatmap
fig = plt.figure(figsize=(12, 8))
sns.heatmap(transaction_volume_pivot, cmap='YlGnBu', fmt=".1f", linewidths=.5, cbar_kws={'label': 'Total Transaction Price'})
plt.xlabel('Hour of Day')
plt.ylabel('Day of Week')
plt.title('Transaction Volume by Day of Week and Hour')
plt.tight_layout()

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Heatmap of Transaction Volume by Day and Hour"
display.display(display.Markdown(F"![{alt}]({image})"))
plt.close(fig)

# Identify top 3 day-hour combinations
top_3_combinations = transaction_volume.sort_values(by='Price', ascending=False).head(3)
print("\nTop 3 Day-Hour Combinations by Total Transaction Amount:")
print(top_3_combinations.to_string(index=False))
````
